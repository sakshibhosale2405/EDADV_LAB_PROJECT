```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('flipkart_cleanedfinal.csv')
sns.set(style="whitegrid")

top_10_brands = df['brand'].value_counts().nlargest(10).index

plt.figure(figsize=(10,6))
sns.countplot(y='brand', data=df[df['brand'].isin(top_10_brands)],
              order=top_10_brands, palette='viridis')
plt.title('Top 10 Brands by Number of Products')
plt.show()
plt.show()
```

```python
plt.figure(figsize=(10,6))

avg_price_display = df.groupby('display_type')['current_price'].mean().sort_values(ascending=False).index

sns.barplot(
    x='display_type',
    y='current_price',
    data=df,
    order=avg_price_display,
    palette='magma',
    errorbar=None      #  Removes the vertical line (error bar)
)

plt.title('Average Price by Display Type')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your cleaned dataset
df = pd.read_csv("flipkart_cleanedfinal.csv")

plt.figure(figsize=(8,8))

# Count display types
display_counts = df['display_type'].value_counts()

# Top 4 categories
top_display = display_counts.nlargest(4)

# Combine remaining into Others
others = display_counts.iloc[4:].sum()

# FIX: Use concat instead of append
final_display = pd.concat([top_display, pd.Series({'Others': others})])

# Colors you can customize
colors = ['#636EFA', '#EF553B', '#00CC96', '#AB63FA', '#FFA15A']

# Pie chart
plt.pie(
    final_display,
    labels=final_display.index,
    autopct='%1.1f%%',
    colors=colors[:len(final_display)],
    startangle=140
)

plt.title("Share of Display Types (Grouped)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))

# Count brands
brand_counts = df['brand'].value_counts()

# Top 5 brands
top5 = brand_counts.nlargest(5)

# Sum of remaining brands
others = brand_counts.iloc[5:].sum()

# FIX: Use concat instead of append
final_brands = pd.concat([top5, pd.Series({'Others': others})])

# Colors
colors = ['#636EFA', '#EF553B', '#00CC96', '#AB63FA', '#FFA15A', '#19D3F3']

# Pie chart
plt.pie(
    final_brands,
    labels=final_brands.index,
    autopct='%1.1f%%',
    colors=colors[:len(final_brands)],
    startangle=140
)

plt.title("Market Share of Top 5 Brands")
plt.show()
```

```python
plt.figure(figsize=(10,10))

launch_counts = df['launch_year'].value_counts().sort_index()

# Keep top 5 years only
top_years = launch_counts.nlargest(5)

# Group all remaining years
others = launch_counts.iloc[5:].sum()

# FINAL dataset for the pie chart
final_years = pd.concat([top_years, pd.Series({'Others': others})])

# Colors matching the number of slices
colors = ['#636EFA', '#EF553B', '#00CC96', '#AB63FA', '#FFA15A', '#19D3F3']

plt.pie(
    final_years,
    labels=final_years.index.astype(str),
    autopct='%1.1f%%',
    colors=colors[:len(final_years)],
    startangle=140
)

plt.title("Product Launch Share by Year (Grouped)")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("flipkart_cleanedfinal.csv")

# Select only numeric columns  includes encoded features automatically
num_df = df.select_dtypes(include=['int64', 'float64'])

# Compute correlation
corr = num_df.corr()

# Plot heatmap
plt.figure(figsize=(16,12))

sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.6,
    cbar=True,
    square=True
)

plt.title("Correlation Heatmap: All Numerical + Encoded (ML-Ready Dataset)", 
          fontsize=16, weight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()

plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Load dataset
df = pd.read_csv("flipkart_cleanedfinal.csv")

# Copy to avoid modifying original
ml_df = df.copy()

# Automatically detect categorical columns
cat_cols = ml_df.select_dtypes(include=['object']).columns

# Create a label encoder
le = LabelEncoder()

# Encode all categorical columns
for col in cat_cols:
    ml_df[col + "_encoded"] = le.fit_transform(ml_df[col].astype(str))

# Drop original categorical columns (optional)
# ml_df = ml_df.drop(columns=cat_cols)

ml_df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Pivot for stacked bar ---
pivot_data = df.pivot_table(
    index='brand',
    columns='screen_resolution',
    aggfunc='size',
    fill_value=0
)

top10_brands = df['brand'].value_counts().nlargest(10).index
pivot_top10 = pivot_data.loc[top10_brands]

# --- Plot ---
pivot_top10.plot(
    kind='bar',
    stacked=True,
    figsize=(14,7),
    colormap='tab20'
)

plt.title("Brand vs Resolution Mix (Stacked Bar)")
plt.xlabel("Brand")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Categorize discount
df['discount_group'] = pd.cut(
    df['discount_percent'],
    bins=[0, 20, 50, 100],
    labels=['Low (<20%)', 'Medium (20-50%)', 'High (>50%)']
)

discount_counts = df['discount_group'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(
    discount_counts,
    labels=discount_counts.index,
    autopct="%1.1f%%",
    colors=['#4e79a7', '#f28e2b', '#e15759'],
    startangle=140
)
plt.title("Discount Level Distribution")
plt.show()
```

```python
heat_data = df.pivot_table(
    index='brand',
    columns='screen_size_inch',
    aggfunc='size',
    fill_value=0
)

plt.figure(figsize=(15,8))
sns.heatmap(
    heat_data,
    cmap='Blues',
    linewidths=0.5
)
plt.title("Brand vs Screen Size Availability (Inventory Heatmap)")
plt.xlabel("Screen Size (inches)")
plt.ylabel("Brand")
plt.show()
```
